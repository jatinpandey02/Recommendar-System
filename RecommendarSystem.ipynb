###Create similarity cross user table

```
from sklearn.metrics.pairwise import cosine_similarity
similarity = pd.DataFrame(cosine_similarity(rating),columns=users['_id'])
similarity.set_index(users['_id'],inplace=True)
```

```
similarity
```

### Vectorize category/tags of posts

```
tags = set()
for i in posts.index:
    if type(posts['category'][i])==float:
        continue
    tags.update(posts['category'][i].split('|'))
```

```
token = dict()
for i,t in enumerate(tags):
    token[t] = i
```

```
token
```

```
posts['token'] = [[0]*len(tags)]*len(posts)
for i in posts.index:
    if type(posts['category'][i]) == float:
        continue
    a = [0]*len(tags)
    for s in posts['category'][i].split('|'):
        a[token[s]] = 1
    posts['token'][i] = a
```

```
posts
```

###Misc

```
rating_T = rating.T
```

```
posts.set_index('_id',inplace=True)
```

###Function to return post ids by order of scores
Scores include both simialirty of tags of already viewed posts and similarity cross table of user

```
def get_posts(user):
    viewed = set()
    for i in rating_T.index:
        if rating_T[user][i] == 1:
            viewed.add(i)
    #print(viewed)
    posts_rec = dict()
    for i in posts.index:
        if i in viewed:
            continue
        score = 0
        div = 0
        for j in similarity[user].index:
            if similarity[user][j] == 1:
                continue
            score+= similarity[user][j]*rating_T[j][i]
            div+= abs(similarity[user][j])

        score = score/div
        #print(score,end=' ')

        for j in viewed:
            a = np.dot(posts['token'][j],posts['token'][i])/(np.linalg.norm(posts['token'][j])*np.linalg.norm(posts['token'][i]))
            if np.isnan(a):
                continue
            score+=a
        
        #print(score)
        posts_rec[i] = score

    return sorted(posts_rec,reverse=True)
```
